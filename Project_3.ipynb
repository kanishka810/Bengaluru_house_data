## **Matplotlib**

Line Graph

```python
import matplotlib.pyplot as plt
plt.plot(df['total_sqft'].head(10),df['price'].head(10),marker = "o",color = "r",linestyle = "-")
plt.xlabel("Total Square Feet")
plt.ylabel("Price")
plt.title("Total Square Feet vs Price")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Bar Graph

```python
avg_price = df.groupby('area')['price'].mean()
plt.bar(avg_price.index,avg_price.values,color = "teal")
plt.xlabel("Area")
plt.ylabel("Average Price")
plt.title("Average Price by Area")
plt.grid(True)
plt.show()
```

Pie Chart

```python
area_counts = df['area'].value_counts()
plt.pie(area_counts,labels = area_counts.index,autopct='%1.1f%%', startangle=90,colors = ["skyblue","lightgreen"])
plt.title("Area Distribution")
plt.show()
```

Donut Chart

```python
society_counts = df['society'].value_counts().head(8)
plt.pie(society_counts,labels = society_counts.index,autopct='%1.1f%%')
plt.gca().add_artist(plt.Circle((0,0), 0.5, color='white'))
plt.title("Society Distribution")
plt.show()
```

Histogram

```python
plt.hist(df['price'],bins = 15,color = 'purple',edgecolor = 'black')
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Price Distribution")
plt.grid(True)
plt.show()
```

Scatter plot

```python
plt.scatter(df['total_sqft'].head(10),df['price'].head(10),color = "blue",alpha = 0.6)
plt.xlabel("Total Square Feet")
plt.ylabel("Price")
plt.title("Total Square Feet vs Price")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Box Plot

```python
df.boxplot(column = 'price',by = 'area')
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Price Distribution by Area")
plt.suptitle("")
plt.show()
```
